## Drop duplicated values

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df.duplicated().sum()
```

## Handle null values

```python
df.isnull().sum()
```

- index, don't have meaning in visualize.
- New and PendingS have null values same as the values of the data so i will drop it.

```python
df.drop(['index', 'New', 'PendingS'], axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['currency'].isnull()]
```

```python
currency_null = df[df['currency'].isnull()]
amount_null = df[df['Amount'].isnull()]

currency_null.equals(amount_null)
```

```python
df['currency'].value_counts().to_frame()
```

- Can see that when the currency is null then the amount is also null.
